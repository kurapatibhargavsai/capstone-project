# Adding distance dummy variable from from_lat,from_long,to_lat,to_long

```python
def haversine(lat1, lon1, lat2, lon2):
     
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    r = 6371  # Radius 
    return c * r

# Calculate distances and add a new column to the dataset
dataset_2['distance'] = haversine(dataset_2['from_lat'], dataset_2['from_long'],
                                dataset_2['to_lat'], dataset_2['to_long'])
```
